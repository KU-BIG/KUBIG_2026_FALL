```python
# 행렬 연산

A = np.array([[1, 2, 3], [4, 5, 6]])

B = np.array([[7, 8], [9, 10], [11, 12]])

# 행렬의 내적 구하기(행렬곱)
dot_product = np.dot(A, B)
print(dot_product)

# 전치 행렬 구하기
print(np.transpose(A))
```

```python
# 행렬곱 함수식을 만들어봤습니다.

import numpy as np

def dot(ar1, ar2):
    if ar1.shape[1] != ar2.shape[0]:
        print("행렬의 크기가 맞지 않습니다.")
        return None
    answer_row = ar1.shape[0]
    answer_col = ar2.shape[1]
    answer = np.zeros((answer_row, answer_col))

    for i in range(answer_row):
        for j in range(answer_col):
            for k in range(ar1.shape[1]):
                answer[i][j] = answer[i][j] + ar1[i][k] * ar2[k][j]

    return answer
```

```python
# 전치 행렬 함수식을 만들어봤습니다.

import numpy as np

def transpose(ar):
    input_row = ar.shape[0]
    input_col = ar.shape[1]

    answer = np.zeros((input_col, input_row))

    for i in range(input_row):
        for j in range(input_col):
            answer[j][i] = ar[i][j]

    return answer
```

```python
# Q4. matplotlib, seaborn의 다양한 기능을 이용하여 더 많은 EDA를 수행해 봅시다.
```

EDA(Exploratory Data Analysis, 탐색적 데이터 분석)는 수집한 데이터를 본격적으로 분석하기 전, 다양한 각도에서 관찰하고 핵심 특성을 파악하는 초기 분석 단계입니다. 데이터의 패턴, 숨은 구조, 이상치(Outlier) 등을 발견하여 데이터에 대한 이해도를 높이는 데 사용됩니다.


```python
# 상관관계 분석

import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

petal_correlation = iris["petal_length"].corr(iris["petal_width"])
sepal_correlation = iris["sepal_length"].corr(iris["sepal_width"])

print(f"petal length-width 상관계수: {petal_correlation:.3f}")
print(f"sepal length-width 상관계수: {sepal_correlation:.3f}")


figure, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.regplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    ax=axes[0],
)
axes[0].set_title(f"Petal correlation: {petal_correlation:.3f}")

sns.regplot(
    data=iris,
    x="sepal_length",
    y="sepal_width",
    ax=axes[1],
)
axes[1].set_title(f"Sepal correlation: {sepal_correlation:.3f}")

plt.tight_layout()
plt.show()
```

# 결과해석

length-width의 상관관계는 sepal에서보다 petal에서 더 두드러집니다. 그 이유는 각 기관의 주요 역할로 추측해볼 수 있습니다.


Petal(꽃잎): 꽃잎은 곤충이나 새와 같은 수분 매개체를 유인하는 기관입니다. 수분 매개체가 잘 앉을 수 있도록 특정 형태적 비율(예: 길쭉한 모양, 넓은 면적 등)을 유지해야 생존 및 번식에 유리합니다. 따라서 진화 과정에서 길이와 너비가 함께 비례하여 커지거나 작아지도록 선택압을 받았을 것입니다.<br>
Sepal(꽃받침): 꽃받침의 주된 역할은 꽃이 피기 전 꽃봉오리를 보호하는 것입니다. 개화 후에는 그 기능이 상대적으로 줄어들기 때문에, 꽃잎에 비해 크기나 비율이 유기적으로 엄격하게 연동될 필요가 적습니다.


```python
# 상자그림(boxplot) - 품종에 따른 petal/sepal 길이 비율

iris["petal_sepal_ratio"] = iris["petal_length"] / iris["sepal_length"]

ratio_mean = iris.groupby("species")["petal_sepal_ratio"].mean()

print("\n품종별 petal/sepal 길이 비율의 평균:")
print(ratio_mean)

sns.boxplot(
    data=iris,
    x="species",
    y="petal_sepal_ratio",
)
plt.title("Petal length / Sepal length by species")
plt.ylabel("Petal-to-sepal length ratio")
plt.tight_layout()
plt.show()
```

### Versicolor는 Setosa와 Virginica의 자연 교배를 통해 태어난 합성 신종(종간 잡종)이다. 그렇다면 형태적인 특징도 두 기존의 종의 사이일까?
두 기존의 종의 중간값과 두 종의 교배를 통해 태어난 신종의 특징을 비교해보겠다.

```python
features = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
]

species_mean = iris.groupby("species")[features].mean()

# setosa와 virginica 평균의 중간값
predicted_midpoint = (
    species_mean.loc["setosa"] + species_mean.loc["virginica"]
) / 2

comparison = pd.DataFrame({
    "setosa": species_mean.loc["setosa"],
    "predicted_midpoint": predicted_midpoint,
    "versicolor": species_mean.loc["versicolor"],
    "virginica": species_mean.loc["virginica"],
})

comparison["midpoint_error"] = abs(
    comparison["versicolor"] - comparison["predicted_midpoint"]
)

print(comparison.round(3))

comparison[["predicted_midpoint", "versicolor"]].plot(
    kind="bar",
    figsize=(10, 5),
)

plt.title("Predicted midpoint vs actual versicolor")
plt.ylabel("Length or width (cm)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
