## Numpy

Numpy의 ndarray를 이용해 다차원 배열을 쉽게 생성하고, 다양한 연산을 수행할 수 있습니다.

```python
# 넘파이 배열(행렬) 생성하기

x = np.array([1, 2, 3])
y = np.array([2, 3, 4])
z = np.array([[1, 2, 3], [2, 3, 4]])

w = np.array([[1, 2, 3]])
```

```python
# Q1. 배열의 타입과 형태, 차원을 각각 구해주세요.

arrays = {"x": x, "y": y, "z": z, "w": w}

for name, array in arrays.items():
    print(f"{name} : "
          f"type = {type(array)}, "     # type = 파이썬 객체의 타입
          f"dtype = {array.dtype}, "    # dtype = 배열 안 원소의 자료형
          f"shape = {array.shape}, "    # shape = 배열의 형태 (행, 열)
          f"ndim = {array.ndim}"        # ndim = 배열의 차원
    )
```

| 배열 | 타입 | 형태(shape) | 차원(ndim) |  |
|---|---|---|---:|---|
| x | numpy.ndarray | (3,) | 1 | 원소 3개를 가진 1차원 배열 |
| y | numpy.ndarray | (3,) | 1 | 원소 3개를 가진 1차원 배열 |
| z | numpy.ndarray | (2, 3) | 2 | 2행 3열의 2차원 배열 |
| w | numpy.ndarray | (1, 3) | 2 | 1행 3열의 2차원 배열 |

```python
# 원소끼리의 연산 수행

print(x + y)
print(x - y)
print(x * y)  # 행렬 곱이 아니라 같은 위치의 원소끼리 곱셈

x1 = np.array([1, 2, 3])
y1 = np.array([1, 2, 3, 4])

# 배열의 크기가 같은 경우에만 연산을 적용할 수 있습니다.
x1 + y1
```

```python
# 리스트를 ndarray로 쉽게 변경할 수 있습니다.

lst1 = [1, 2, 3]
lst2 = [4, 5, 6]
lst3 = [7, 8, 9]
print(type(lst1))

arr = np.array([lst1, lst2, lst3])
print(type(arr))

arr
```

```python
# numpy 내의 데이터값은 숫자, 문자열, 불 값 등 모두 가능합니다.
# 한 개의 ndarray 객체에 int와 float가 함께 있을 수 없습니다.

print(np.array([1, 2, 3]).dtype)
print(np.array([1, 2, 3.]).dtype)   # 정수와 실수를 함께 입력하면 실수형으로 변환

print(np.array([1, 2, 'test']).dtype)
```

```python
# reshape() 함수를 이용해 형태나 차원의 크기를 변경합니다.

array1 = np.zeros(9, dtype = 'int32')
array1.reshape(3, 3)
print(array1)                    # 정수0을 9개 가진 1차원 배열
print(array1.reshape(3, 3))      # 정수0을 3행 3열 가진 2차원 배열

array2 = np.arange(10)           # 0부터 9까지의 정수 생성 (10개)
print(array2.reshape(-1, 5))     # -1 : 행 크기는 자동 계산 , 5 : 5열 (2행 5열)
print(array2.reshape(5, -1))     # 5 : 5행 , -1 : 열 크기는 자동 계산 (5행 2열)

array3 = np.arange(8).reshape(2, 2, 2) # 0부터 7까지의 정수 생성, 3차원 배열로 변환
print(array3.tolist())                 # Numpy 배열을 리스트로 변환
```

```python
# 넘파이 인덱싱

array2 = array2.reshape(-1,)  # -1 : 1차원으로 펼치기
print(array2)

print(array2[3])   # 파이썬 인덱스는 0부터 시작, [3] : 4번째 원소
print(array2[3:5]) #  슬라이싱 : 인덱스 3부터 5-1까지 = 인덱스 3~4

print(array2[-1], array2[-2]) # 음수 인덱스 : -1은 마지막 원소, -2는 뒤에서 두 번째 원소
```

```python
# ndarray 내의 데이터 값을 간단히 수정할 수 있습니다.

array2[3] = 7  # 인덱스3 원소를 7로 변경
array2[7] = 3  # 인덱스7 원소를 3으로 변경

print(array2)
```

```python
# Q2. array2에서 데이터 값이 4보다 큰 데이터만 추출해 주세요. (2가지 이상의 방법 이용)

print(array2[array2 > 4])            # 방법1 : 조건식을 이용한 인덱싱
print(array2[np.where(array2 > 4)])  # 방법2 : np.where() = 조건을 만족하는 값의 위치 반환
```

```python
# 정렬

score = np.array([78, 95, 84, 98, 88])

score_sorted = np.sort(score)  # 오름차순 정렬
print(score_sorted)

# 내림차순 정렬 : np.sort()로 정렬 후, [::-1]을 이용해 역순으로 뒤집기
score_sorted_desc = np.sort(score)[::-1]
print(score_sorted_desc)


score2 = np.array([[3, 2, 1], [6, 5, 4]])

sorted_by_row = np.sort(score2, axis=1)     # 각 행 내부 오름차순 정렬
print(sorted_by_row)

sorted_by_column = np.sort(score2, axis=0)  # 각 열 내부 오름차순 정렬
print(sorted_by_column)
```

```python
# 행렬 연산

A = np.array([[1, 2, 3], [4, 5, 6]])      # 2x3 행렬

B = np.array([[7, 8], [9, 10], [11, 12]]) # 3x2 행렬

# 행렬의 내적 구하기(행렬곱)
dot_product = np.dot(A, B)                # 2x3 행렬과 3x2 행렬 곱 = 2x2 행렬
print(dot_product)

# 전치 행렬 구하기
print(np.transpose(A))                    # 2x3 행렬 A의 Transpose = 3x2 행렬
```

```python
data2 = pd.DataFrame({'Planet': ['Earth', 'Mars', 'Jupyter', 'Saturn', 'Neptune'],
          'Moons': [1, 2, 79, 83, 14],
          'Diameter_km':[12472, 6779, 139820, 116460, 49244],
          'Has_Rings': [False, False, True, True, True]})
```

```python
# 데이터 선택과 필터링
print(data2['Moons'])  # 'Moons'열만

# 고리가 있는 행성만 선택
print(data2[data2["Has_Rings"] == True])  # 'Has_Rings'열이 True인 행만 

# 새로운 열 추가
data2["Moons_Per_Thousand"] = data2["Moons"] / (data2["Diameter_km"] / 1000)  # 지름 1,000km당 위성 수 : 새로운 열로 추가

print(data2)
```

```python
# iloc 와 loc

# iloc은 행과 열의 위치(숫자 인덱스)를 기준으로 데이터를 선택합니다.
print(data2.iloc[2])         # 3번째 행 선택
print(data2.iloc[2, 1])      # 3번째 행, 2번째 열 선택
print(data2.iloc[1:4, 1:3])  # 2-4번째 행, 2-3번째 열 선택

# loc는 행과 열의 이름(라벨)을 기준으로 데이터를 선택합니다.
print(data2.loc[2])                         # 인덱스 라벨이 2인 행
print(data2.loc[2, "Moons"])                # 인덱스 라벨이 2인 행의 'Moons' 열 값
print(data2.loc[1:3, ["Planet", "Moons"]])  # 인덱스 라벨이 1-3인 행과 'Planet', 'Moons' 열 선택
                                            # loc 슬라이싱은 끝 인덱스 라벨도 포함 

# Q3. loc와 iloc를 이용해 다음 문제를 풀어봅시다.

# 지름이 50,000km 이상인 행성의 이름과 위성 수 구하기

large_planets = data2.loc[data2["Diameter_km"] >= 50000, ["Planet", "Moons"]]
print(large_planets)

# 처음 3행과 2-3번째 열 구하기

print(data2.iloc[:3, 1:3])
```

```python
# 데이터 정렬

data3 = data2.sort_values(by="Moons", ascending=False)  
# 'Moons'열 기준 내림차순(ascending=False) 정렬 : 위성 수가 많은 순으로 정렬
print(data3)

data3.sort_index()
# 인덱스 기준으로 다시 오름차순 정렬
```

# 시각화

```python
# 히스토그램
iris.hist(figsize = (10, 10))
```

```python
# 산점도
sns.scatterplot(x='petal_length', y='petal_width', data=iris)
# x축 : 꽃잎 길이, y축 : 꽃잎 너비
```

```python
# 그룹 산점도 - matplotlib
groups = iris.groupby('species') # iris데이터를 종별로 나누기 (setosa, versicolor, virginica)
fig, ax = plt.subplots() # 중요
for name, group in groups:
  ax.plot(group.petal_length,
          group.petal_width,
          marker='o',
          linestyle='', label=name)
ax.legend(fontsize=12, loc='upper left') # legend position
plt.title('Scatter Plot of iris by matplotlib', fontsize=20)
plt.xlabel('Petal Length', fontsize=14)
plt.ylabel('Petal Width', fontsize=14)
plt.show()
```

```python
# 그룹 산점도 - seaborn
sns.scatterplot(x="petal_length", y="petal_width", data=iris, hue="species")
# hue="species" : 종별로 색상 구분
```

```python
# Q4. matplotlib, seaborn의 다양한 기능을 이용하여 더 많은 EDA를 수행해 봅시다.
```

```python
# 상관관계 분석

numeric_iris = iris.select_dtypes(include="number")  # 수치형 데이터만 선택

correlation_matrix = numeric_iris.corr()  # 상관행렬 계산
print(correlation_matrix)

plt.figure(figsize=(10,8))

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
# annot=True : 각 셀에 상관계수 표시, fmt=".2f" : 소수점 2자리까지 표시
# cmap="coolwarm" : 상관계수에 따라 색을 다르게 표시, vmin=-1/vmax=1 : 색상 범위를 -1~1로 설정

plt.title("Correlation Matrix of Iris Features")
```

```python
# 상자그림(boxplot) : 종별 꽃잎 길이 분포

plt.figure(figsize=(10,8))

sns.boxplot(x="species", y="petal_length", data=iris, hue="species")
plt.title("Boxplot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")

```

기타 자유롭게 자신만의 EDA를 수행해 주세요 (optional)

```python
# pairplot : 여러 수치형 변수 간의 관계를 한 번에 시각화

sns.pairplot(iris, hue="species", diag_kind="hist")
```

```python
# stripplot : 종별 꽃잎 길이 관측치 분포

plt.figure(figsize=(10, 8))

sns.stripplot(data=iris, x="species", y="petal_length", hue="species")

plt.title("Petal Length Observations by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
```

```python
# violinplot : 종별 꽃잎 길이 분포와 밀도

plt.figure(figsize=(10, 8))

sns.violinplot(data=iris, x="species", y="petal_length", hue="species")
plt.title("Violin Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
```
